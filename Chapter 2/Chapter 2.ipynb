```r
set.seed(2021)
myseries <- aus_retail %>%
  filter(`Series ID` == sample(aus_retail$`Series ID`,1))
```

`autoplot()`:

```r
autoplot(myseries, Turnover)
```

## Trend

There is definitely an upward Trend in this time series.

`gg_season()`:

```r
gg_season(myseries, Turnover, labels = "both")
```

## Seasonality

The seasonal plot shown above exhibits clear seasonal behaviour. Every December, there is a sharp rise in `Turnover` followed by a steep fall in January.

## Cyclicity

It looks like there may be some cyclic behaviour observed during the years 2000-2001. It would appear that the cycle begins around April of 2000 with an unusual (up until that point) rise in `Turnover` followed by a steep drop then a slow rise for the rest of the year. Then in the year 2001 we observe that the increase in `Turnover` for this year is much larger relative to the previous years. This cyclic behaviour conforms well to the general rule that buisiness cycles tend to last around two years and that the duration of the cycle lasts longer than the seasonal pattern (which in this case is one year).

`gg_subseries()`:

```r
gg_subseries(myseries, Turnover)
```

`gg_lag()`:

```r
gg_lag(myseries, Turnover, geom = "point")
```

`ACF()`:

```r
myseries %>%
    ACF(Turnover) %>%
    autoplot()
```

## What I've Learned

I've learned that `Turnover` is increasing in Australia each year, that it is strongly seasonal with most `Turnover` in December. I've also learned that `Turnover` in Australia has been shown to be affected by the buisiness cycle.

```r
pigs <- aus_livestock %>%
    filter(Animal == "Pigs", State == "Victoria")
```

```r
pigs_90_95 <- pigs %>%
    filter(year(Month) >= 1990 & year(Month) <= 1995)
```

## 1990-1995:

`autoplot()`:

```r
autoplot(pigs_90_95, Count)
```

`ACF()`:

```r
pigs_90_95 %>%
    ACF(Count) %>%
    autoplot()
```

Because at least one of the spikes in ACF plot, supercedes the line of 5% significance. We have sufficient evidence that the number of pigs slaughtered in Victoria between the years 1990 and 1995 is not white noise.

## 1972-2018:

In this section, we will create the same two graphs, but this time for the entire time period of the data. We will then compare it to white noise.

`autoplot()`:

```r
autoplot(pigs, Count)
```

```r
pigs %>%
    ACF(Count) %>%
    autoplot()
```

When we incorporate the full dataset, there is even more evidence that the data does not constitute white noise. 

In using a longer time-period, the autocorrelations have increased.

## a.

```r
dgoog <- gafa_stock %>%
  filter(Symbol == "GOOG", year(Date) >= 2018) %>%
  mutate(trading_day = row_number()) %>%
  update_tsibble(index = trading_day, regular = TRUE) %>%
  mutate(diff = difference(Close))
```

## b.

If we want to look at the difference's ACF, there cannot be any gaps in the data. Google trading days never occur on weekends so we would need to get rid of the gaps. This is accomplished by creating a new tsibble index 'trading_day'.

## c.

```r
autoplot(dgoog, diff)
```

```r
dgoog %>%
    ACF(diff) %>%
    autoplot()
```

## d.

While one of the autocorrelation spikes does surpass the significance lines in the ACF plot, I believe that to reject the null hypothesis that this data constitutes white noise would be type 1 error.

In other words, this data looks like white noise to me.
